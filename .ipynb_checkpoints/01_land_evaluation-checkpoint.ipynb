# Assessment of Renewable Energy Potentials based on Land Availability

The analysis of land availability for renewable energy potentials is divided into three stages. In this script, we will focus on the first stage: **Land Evaluation**.

-Goal: find out available lands that can be used for renewable energy development.

-Steps:

1. Import data
2. Data unification
3. Calculate the total area and plot the map
4. Merge the layers

## 1. Import data

a. Import the **basemap** which is the **administrative level in Vietnam (VN)**. Here we use **level 2** to proceed.

b. Import the **map layers** that are occupied and not suitable for renewable energy development. Here the layers include: 

(a) Landuse Map (residential areas, industrial areas, military areas, etc.) <br>
(b) Building Map (building outline from a bird's eye view)<br>
(c) Water Areas Map (Rivers, Lakes, Reservoirs, Docks, Glaciers, etc.) <br>
(d) Roads Map (road networks, highways, etc)<br>
(e) Railways Map (Railways, Subways, Trams, Lifts, and Cable Cars) <br>
(f) Transport Stations Map (railway stations, subway stations, airports, aprons) <br>
*The descriptions of the categories can be found in [OpenStreetMap](http://download.geofabrik.de/osm-data-in-gis-formats-free.pdf)


```python
# gdf: geodataframe
adminmap = os.path.abspath('Vietnam/00_data_vector/01_admin_level2/vnm_admbnda_adm2_gov_20201027.shp')
%time adminmap_gdf = geopandas.read_file(adminmap)
```

```python
# (a) Landuse (forests, parks, residential, industrial, military, etc.) (polygon)
land = os.path.abspath('Vietnam/00_data_vector/02_occupied_land/gis_osm_landuse_a_free_1.shp')
%time land_gdf = geopandas.read_file(land)

# (b) Building (polygon)
building = os.path.abspath('Vietnam/00_data_vector/03_buildings/gis_osm_buildings_a_free_1.shp')
%time building_gdf = geopandas.read_file(building)

# (c) Water Areas Map (Rivers, Lakes, Reservoirs, Docks, Glaciers, etc.)
water = os.path.abspath('Vietnam/00_data_vector/04_water_area/gis_osm_water_a_free_1.shp')
%time water_gdf = geopandas.read_file(water)

# (d) Roads Map (road networks, highways, etc) (line)
roads = os.path.abspath('Vietnam/00_data_vector/05_roads/gis_osm_roads_free_1.shp')
%time roads_gdf = geopandas.read_file(roads)

# (e) Railways Map (Line) (Railways, Subways, Trams, Lifts, and Cable Cars)
rails = os.path.abspath('Vietnam/00_data_vector/06_railways/gis_osm_railways_free_1.shp')
%time rails_gdf = geopandas.read_file(rails)

# (f) Transport Stations Map (railway stations, subway stations, airports, aprons)
transport = os.path.abspath('Vietnam/00_data_vector/07_transport_stations/gis_osm_transport_free_1.shp')
%time transport_gdf = geopandas.read_file(transport)
```

## 2. Data Unification

### a. Sort and re-index the **basemap** (**administrative level in Vietnam (VN)**)

```python
print(adminmap_gdf.columns)
```

```python
# drop columns that are not needed for calculations
map_drop_gdf = adminmap_gdf.drop(columns=['Shape_Leng','Shape_Area', 'ADM1_VI', 'ADM2_VI', 'ADM2_PCODE', 'ADM2_REF', 
                                           'ADM2ALT1EN', 'ADM2ALT2EN', 'ADM2ALT1VI','ADM2ALT2VI','ADM1_PCODE','ADM0_EN','ADM0_VI','ADM0_PCODE',
                                           'date','validOn','validTo'])
#rename the titles
# create a dictionary and renew the titles
dict = {'NAME_0': 'country_name',
        'NAME_1': 'ad_lv_1',
        'VARNAME_2': 'ad_lv_2',
        'geometry': 'geometry'}
  
# call rename () method
map_drop_gdf.rename(columns=dict,
          inplace=True)
  
map_drop_gdf.head()
```

### b. Sort and re-index the other map layers.
Only three columns shall be kept: **type/fclass, name, geometry**

```python
# list the titles of each column for selection
titles_array = [land_gdf,building_gdf,water_gdf,roads_gdf,rails_gdf,transport_gdf]

for i in titles_array:
    print(i.columns)
```

```python
# drop columns that are not needed for calculations
land_drop_gdf = land_gdf.drop(columns=['osm_id','code'])
building_drop_gdf = building_gdf.drop(columns=['osm_id','code', 'type'])
water_drop_gdf =water_gdf.drop(columns=['osm_id','code'])

# remove unnecessary columns
roads_drop_gdf = roads_gdf.drop(columns=['osm_id','code','ref','maxspeed','layer','bridge','tunnel', 'oneway'])
# road data needs to be reduced since it is too detailed
# set up a list that need to remove. And remove the names in column 'fclass' that are included in the list.
# drop columns that are not needed for calculations
drop_roads = ['footway','residential','unclassified','living_street','pedestrian','track_grade1','bridleway','path','steps','unknown']
roads_drop_gdf = roads_drop_gdf[~roads_drop_gdf['fclass'].isin(drop_roads)]

# remove unnecessary columns
rails_drop_gdf = rails_gdf.drop(columns=['osm_id','code','layer','bridge','tunnel'])
# rails data needs to be reduced since it is too detailed
# set up a list that need to remove. And remove the names in column 'fclass' that are included in the list.
# drop columns that are not needed for calculations
drop_rails = ['subway','drag_lift','chair_lift','cable_car','gondola','goods','other_lift']
rails_drop_gdf = rails_drop_gdf[~rails_drop_gdf['fclass'].isin(drop_rails)]

# remove unnecessary columns
trans_drop_gdf = transport_gdf.drop(columns=['osm_id','code'])
# transport data needs to be reduced since it is too detailed
# set up a list that need to remove. And remove the names in column 'fclass' that are included in the list.
# drop columns that are not needed for calculations
drop_trans = ['tram_stop','bus_stop','bus_station','taxi_rank','helipad','ferry_terminal','aerialway_station']
trans_drop_gdf = trans_drop_gdf[~trans_drop_gdf['fclass'].isin(drop_trans)]

layers = [land_drop_gdf,building_drop_gdf,water_drop_gdf,roads_drop_gdf,rails_drop_gdf,trans_drop_gdf]

for i in layers:
    print(i.columns)
```

### c. Scale up the **buffer** of each vector

```python
# Polygon file (land use, building, water)
# String file (roads, railways, transportation)
# functions of geopandas package
# increase the buffer to make every layer become polygons (longitude crs. unit in 0.01 degree of longitude (1.11 km))
for layer in layers:
        layer['geometry'] = layer.geometry.buffer(0.01, cap_style = 3, join_style = 1)
```

## 3. Calculate the total area and plot the map

[EPSG:4326/WGS 84](https://epsg.io/4326) - World Geodetic System 1984, used in GPS (common longtitude/latitude system) <br>
[EPSG:3405/VN-2000 / UTM zone 48N](https://epsg.io/3405) - Accurate projection and caluclation of areas (units in meters) <br>
*UTM: Universal Transverse Mercator coordinate system

### a. Calculate the total area of Vietnam via **basemap** (**administrative level in Vietnam (VN)**)

```python
# check the CRS of all layers
layers_crs = [map_drop_gdf, land_drop_gdf,building_drop_gdf,water_drop_gdf,roads_drop_gdf,rails_drop_gdf,trans_drop_gdf]

for i in layers_crs:
    print(i.crs)
```

```python
# crs for meter
# from https://epsg.io/
crs_m = CRS.from_proj4("+proj=utm +zone=48 +ellps=WGS84 +towgs84=-192.873,-39.382,-111.202,-0.00205,-0.0005,0.00335,0.0188 +units=m +no_defs ")
crs_m
```

```python
# Original CRS: EPSG:4326 (Unit: longtitude, latitiude)
# Reprojection to crs
map_m_gdf = map_drop_gdf.to_crs(crs_m)
# Areas in square kilometers
map_km_df = map_m_gdf.area/ 10**6

print('Total area of', name, ':', map_km_df.sum(axis = 0, skipna = True),'km^2')
```

Compared to the [official value of total area (310,070 km^2 in 2018)](https://data.worldbank.org/indicator/AG.LND.TOTL.K2?locations=VN), there are deviations because the data is calculated based on different CRSs, the projections of the map will cause the differences.

## 4. Merge the layers

### a. Merging process
*It may take some time: approx. 15min

```python
# Overlay: basemap minu all occupied land layers  
%time merge1_gdf = geopandas.overlay(map_drop_gdf, rails_drop_gdf, how = 'difference')
%time merge2_gdf = geopandas.overlay(merge1_gdf, trans_drop_gdf, how = 'difference')
%time merge3_gdf = geopandas.overlay(merge2_gdf, water_drop_gdf, how = 'difference')
```

```python
%time merge4_gdf = geopandas.overlay(merge3_gdf, land_drop_gdf, how = 'difference')
%time merge5_gdf = geopandas.overlay(merge4_gdf, building_drop_gdf, how = 'difference')
%time merge6_gdf = geopandas.overlay(merge5_gdf, roads_drop_gdf, how = 'difference')
```

```python
# after completing merging all layers
merge6_gdf
```

```python
# Save the shapefile
merge6_gdf.to_file(r'Vietnam/01_output_stage1/01_vn_available_land.shp')
```

### b. Calculate the total area

```python
# Original CRS: EPSG:4326 (Unit: longtitude, latitiude)
# Reprojection to crs
map_merge_m_gdf = merge6_gdf.to_crs(crs_m)
# Areas in square kilometers
map_merge_km_df = map_merge_m_gdf.area/ 10**6

# Total areas in square kilometers
print('Total area of', name, ':', map_merge_km_df.sum(axis = 0, skipna = True),'km^2')
```

```python
# Plot the data with a modified spatial extent
fig, ax = plt.subplots(figsize=(10, 20))

# Use VN_landuse boundary to define all sizes of the maps
ax.set_xlim(101, 110)
ax.set_ylim(8, 24)

# add basemap to the plot (default color)
merge6_gdf.plot(ax=ax, categorical=True, legend=True)

# Set titles
ax.set_title('Vietnam Available Land for renewables Potentials\n(Total Area :{0:.{1}f}km\u00b2)'.format(map_merge_km_df.sum(axis = 0, skipna = True), 2), fontsize=20)
ax.set_axis_off()

# Save the graph into tiff file
plt.savefig(r'Vietnam/01_output_stage1/01_vn_available_land.tif', dpi=400)
plt.show()
```
